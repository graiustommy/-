````
# 袩褉懈屑械褉 写邪薪薪褘褏 (锌械褉胁褘械 8 褋褌褉芯泻)
import pandas as pd
import numpy as np
from datetime import datetime
rng = pd.date_range(start="2020-01-01", periods=200, freq="D")
t = np.arange(len(rng))
seasonal = 10 * np.sin(2 * np.pi * t / 30)
trend = 0.05 * t
noise = np.random.normal(scale=2.0, size=len(rng))
series = 50 + trend + seasonal + noise
df = pd.DataFrame({'ds': rng, 'y': series})
df.head(8)
````

````
# 袘褘褋褌褉邪褟 胁懈蟹褍邪谢懈蟹邪褑懈褟 (matplotlib)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
plt.plot(df['ds'], df['y'])
plt.title('Synthetic time series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
````

````
# 袥懈薪械泄薪邪褟 褉械谐褉械褋褋懈褟 薪邪 谢邪谐懈
from sklearn.linear_model import LinearRegression
df_ml = df.copy()
for lag in (1,2,3,7):
    df_ml[f'lag_{lag}'] = df_ml['y'].shift(lag)
df_ml = df_ml.dropna()
X = df_ml[[c for c in df_ml.columns if c.startswith('lag_')]]
y = df_ml['y']
model = LinearRegression().fit(X, y)
print('Coef:', model.coef_)
````
