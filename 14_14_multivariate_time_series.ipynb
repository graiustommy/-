Vector autoregression (VAR) and multivariate forecasting basics.

````
# Пример данных (первые 8 строк)
import pandas as pd
import numpy as np
from datetime import datetime
rng = pd.date_range(start="2020-01-01", periods=200, freq="D")
t = np.arange(len(rng))
seasonal = 10 * np.sin(2 * np.pi * t / 30)
trend = 0.05 * t
noise = np.random.normal(scale=2.0, size=len(rng))
series = 50 + trend + seasonal + noise
df = pd.DataFrame({'ds': rng, 'y': series})
df.head(8)
````

````
# Быстрая визуализация (matplotlib)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
plt.plot(df['ds'], df['y'])
plt.title('Synthetic time series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
````

````
# Генерируем второй ряд, демонстрируем VAR
import pandas as pd
from statsmodels.tsa.api import VAR
np.random.seed(1)
x1 = df['y'].values + np.random.normal(scale=1.0, size=len(df))
x2 = 0.5*df['y'].values + np.random.normal(scale=1.5, size=len(df))
df_var = pd.DataFrame({'x1': x1, 'x2': x2})
model = VAR(df_var)
res = model.fit(2)
print(res.summary())
````
