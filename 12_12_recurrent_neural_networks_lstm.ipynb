````
# –Я—А–Є–Љ–µ—А –і–∞–љ–љ—Л—Е (–њ–µ—А–≤—Л–µ 8 —Б—В—А–Њ–Ї)
import pandas as pd
import numpy as np
from datetime import datetime
rng = pd.date_range(start="2020-01-01", periods=200, freq="D")
t = np.arange(len(rng))
seasonal = 10 * np.sin(2 * np.pi * t / 30)
trend = 0.05 * t
noise = np.random.normal(scale=2.0, size=len(rng))
series = 50 + trend + seasonal + noise
df = pd.DataFrame({'ds': rng, 'y': series})
df.head(8)
````

````
# –С—Л—Б—В—А–∞—П –≤–Є–Ј—Г–∞–ї–Є–Ј–∞—Ж–Є—П (matplotlib)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
plt.plot(df['ds'], df['y'])
plt.title('Synthetic time series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
````

````
# LSTM –Љ–Є–љ–Є–Љ–∞–ї—М–љ—Л–є –њ—А–Є–Љ–µ—А (Keras)
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
# Prepare a tiny dataset
data = df['y'].values.reshape(-1,1)
# build simple sequences
X, y = [], []
for i in range(20, 180):
    X.append(data[i-20:i,0])
    y.append(data[i,0])
X = np.array(X).reshape(-1,20,1)
y = np.array(y)
model = Sequential([LSTM(16, input_shape=(20,1)), Dense(1)])
model.compile(optimizer='adam', loss='mse')
print('Starting quick fit (1 epoch)...')
model.fit(X,y,epochs=1,verbose=1)
````
