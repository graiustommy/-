# 09 Feature Engineering For Ml

Preparing time series features for ML models: lags, rolling stats, date features.

**Короткое объяснение:**

- Данные: синтетический ряд с трендом, сезонностью и шумом (200 точек).
- Цель: показать ключевые приёмы для изучаемой темы.

````
# Пример данных (первые 8 строк)
import pandas as pd
import numpy as np
from datetime import datetime
rng = pd.date_range(start="2020-01-01", periods=200, freq="D")
t = np.arange(len(rng))
seasonal = 10 * np.sin(2 * np.pi * t / 30)
trend = 0.05 * t
noise = np.random.normal(scale=2.0, size=len(rng))
series = 50 + trend + seasonal + noise
df = pd.DataFrame({'ds': rng, 'y': series})
df.head(8)
````

````
# Быстрая визуализация (matplotlib)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
plt.plot(df['ds'], df['y'])
plt.title('Synthetic time series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
````

````
# Простая техника создания лагов и скользящих средних
df_feat = df.copy()
for lag in (1,2,3,7,14):
    df_feat[f'lag_{lag}'] = df_feat['y'].shift(lag)
df_feat['rolling_7'] = df_feat['y'].rolling(7).mean()
df_feat.dropna().head()
````
